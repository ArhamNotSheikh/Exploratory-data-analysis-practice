STEP 2: DATA PREPRATION



*   Dropping irrelevent colums and rows
* identifying duplicated columns
*  Renaming colums
*   Feature creation






```python
df = df[['coaster_name',
    'Location', 'Status',
    'Manufacturer',
    'year_introduced',
    'latitude', 'longitude', 'Type_Main',
    'opening_date_clean',
    'speed_mph',
    'height_ft',
    'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.dtypes
```

```python
df['opening_date_clean']=pd.to_datetime(df['opening_date_clean'])
```

#Learning to rename our coloms;

```python
df.rename(columns={'coaster_name':'Coaster_Name',
                   'year_introduced':'Year_Intro'})
```

To check missing or null values

```python
df.isna().sum()
#sum of null values per colom
```

```python
df.duplicated()
#showing rownumber and if duplicated
```

```python
df.loc[df.duplicated()]
#shows the whole row which is(if it is) duplicated
```

```python
#we can run this duplicate command on spesific rows/colloms

df.duplicated(subset=['coaster_name'])
```

```python
df.loc[df.duplicated(subset=['coaster_name'])]
```

```python
#we can see there are duplicates now lets see what are they why etc by query
df.query('coaster_name=="Crystal Beach Cyclone"')
```

```python
df.columns
```

```python
df = df.loc[~df.duplicated(subset=['coaster_name', 'Location', 'opening_date_clean'])].reset_index(drop=True)
```

```python
ax=df['year_introduced'].value_counts()\
  .head(10)\
  .plot(kind='bar',title='Top 10 year coasters Introduces')
ax.set_xlabel('year Introduces')
ax.set_ylabel('Number of coasters')
#counts unique vales
```

```python
ax=df['speed_mph'].plot(kind='hist',bins=20,title='Coaster speen in mph')
ax.set_xlabel('mph')
ax.set_ylabel('Number of coasters')
```

```python
ax=df['speed_mph'].plot(kind='kde',title='Coaster speen in mph')
ax.set_xlabel('mph')
ax.set_ylabel('Number of coasters')
```

ASKING QUESTIONS ABOUT DATA

locations with the fastest coastrs(min 10)

```python
df.query('Location!="other"')\
  .groupby('Location')['speed_mph']\
  .agg(['mean','count'])\
  .query('count>=10')\
  .sort_values('mean',ascending=False)\
  .head(10)

#
```
